```python
###################################### function to create spark dataframe from single file
def make_sparkdf(file_list,file_num):
    
    fs = gcsfs.GCSFileSystem(project='deft-return-346819')
    
    dtype={"userid": int, "username": "string", "acctdesc":"string","location":"string","following":int,"followers":int,
       "totaltweets":int,"usercreatedts":"string","tweetid":int,"tweetcreatedts":"string","retweetcount":int,"text":"string",
       "hashtags":"string","language":"string","coordinates":"string","favorite_count":int,"extractedts":"string"}
    
    df_pd = pd.read_csv(fs.open(file_list[file_num]), compression = 'gzip', index_col = 0)#,dtype=dtype)
    
    schema = StructType([
    StructField("userid", LongType(), True),    
    StructField("username", StringType(), True),
    StructField("acctdesc", StringType(), True),
    StructField("location", StringType(), True),
    StructField("following", IntegerType(), True),
    StructField("followers", IntegerType(), True),    
    StructField("totaltweets", IntegerType(), True),
    StructField("usercreatedts", StringType(), True),
    StructField("tweetid", LongType(), True),
    StructField("tweetcreatedts", StringType(), True),
    StructField("retweetcount", IntegerType(),True),
    StructField("text", StringType(), True),
    StructField("hashtags", StringType(),True),
    StructField("language", StringType(), True),
    StructField("coordinates", StringType(), True),
    StructField("favorite_count", IntegerType(), True),
    StructField("extractedts", StringType(),True),
    ])

    df=spark.createDataFrame(df_pd,schema=schema) 
    df.createOrReplaceTempView("df")
    
    df = df.withColumn("usercreatedts",df.usercreatedts.cast(TimestampType()))\
                 .withColumn("tweetcreatedts",df.tweetcreatedts.cast(TimestampType()))\
                 .withColumn("extractedts",df.extractedts.cast(TimestampType()))
    
    df = df.select('username','location','text','language','following','followers','retweetcount','favorite_count','tweetcreatedts')
    df = df.filter(col("language")=='en')
    
    return df
    
```

```python
################################ function to combine spark dataframes
def union_dfs(file_list):
    
    df = make_sparkdf(file_list,0)
    
    for i in range(1,len(file_list)):
        df = df.union(make_sparkdf(file_list,i))
    
    return df
```

```python
r_rdd=df3.select("text").rdd.flatMap(lambda x:x)
```

```python
################################################# functio to obtain tokens from text
stop_words = set(stopwords.words('english'))
table = str.maketrans('', '', string.punctuation)
lmtzr = WordNetLemmatizer() # see https://www.nltk.org/_modules/nltk/stem/wordnet.html for details

def get_tokens(line):
    # get tokens from line
    tokens = word_tokenize(line)
    # convert to lower case
    tokens = [w.lower() for w in tokens]
    # remove punctuations from each word
    stripped = [w.translate(table) for w in tokens]
    # remove remaining tokens that are not alphabetic
    words = [word for word in stripped if word.isalpha()]
    # filter out stop words
    words = [w for w in words if not w in stop_words]
    # lemmatizing the words, see https://en.wikipedia.org/wiski/Lemmatisation
    words = [lmtzr.lemmatize(w) for w in words]
    return (words)

c_rdd = r_rdd.map(lambda line: (1,get_tokens(line)))
```

```python
# remove words of frequency over 3000
doc_stop_words = c_rdd.flatMap(lambda r: r[1]).map(lambda r: (r,1)).reduceByKey(lambda a,b: a+b)
doc_stop_words = doc_stop_words.filter(lambda a: a[1]>3000).map(lambda r: r[0]).collect()

# throw away stop words and words that are just single letters.
c_rdd = c_rdd.map(lambda r: (r[0],[w for w in r[1] if not w in doc_stop_words and not len(w)==1])) 

#c_rdd.take(1)[0][1][:10]
```

```python
raw_df = spark.createDataFrame(c_rdd, ["dummy","topic"])
raw_df.cache()
```

## Defining Functions

```python
def removePattern(inputText, pattern):
    r = re.findall(pattern, inputText)
    for i in r:
        inputText = re.sub(i, '', inputText)        
    return inputText

def cleanTweet(txt):
    '''
    Remove twitter return handles (RT @xxx:)
    '''
    txt = removePattern(txt, 'RT @[\w]*:')
    '''
    Remove twitter handles (@xxx)
    '''
    txt = removePattern(txt, '@[\w]*')
    '''
    Remove URL links (httpxxx)
    '''
    txt = removePattern(txt, 'https?://[A-Za-z0-9./]*')
    '''
    Remove special characters, numbers, punctuations
    '''
    txt = re.sub('[^A-Za-z]+', ' ', txt)
    return txt

def getCleanTweetText(filteredTweetText):
    return ' '.join(filteredTweetText)

def getSentimentScore(tweetText):
    analyzer = SentimentIntensityAnalyzer()
    vs = analyzer.polarity_scores(tweetText)
    return float(vs['compound'])

def getSentiment(score):
     #return 1 if score > 0 else 0
    if score < 0.0:
        return 'Negative'
    elif score == 0.0:
        return 'Neutral'
    else:
        return 'Positive'


def getTweetArray(tweet):
    return tweet.split(' ')
```

### Create graphframe of mentions

```python
################################# define functions to create and visualize graphframes
from graphframes import *

# helper function to filter nodes (vertices) by subject and object, and establish edges
def vertices_edges_split_cond(df, condition):
    sub = df.filter(condition).select("username").withColumnRenamed("username","id")
    obj = df.filter(condition).select("mentioned_user").withColumnRenamed("mentioned_user","id")
    v = sub.union(obj).distinct()
    e = df.filter(condition).select("username","mentioned_user","predicate")\
    .withColumnRenamed("username","src").withColumnRenamed("mentioned_user","dst")
    return v, e

# helper function to filter nodes (vertices) by subject and object, and establish edges
def vertices_edges_split(df):
    sub = df.select("username").withColumnRenamed("username","id")
    obj = df.select("mentioned_user").withColumnRenamed("mentioned_user","id")
    v = sub.union(obj).distinct()
    e = df.select("username","mentioned_user","predicate")\
    .withColumnRenamed("username","src").withColumnRenamed("mentioned_user","dst")
    return v, e

def draw_graphframe(g):
    
    #### Drawing multidigraph
    G = nx.MultiDiGraph()

    for row in g.edges.select('src','dst').take(g.edges.count()):
        G.add_edge(row['src'],row['dst'])

    degrees = sorted(list(G.in_degree),key=lambda x: x[1],reverse=True)
    nodename = [n[0] for n in list(degrees)]
    node_in_deg = [n[1] for n in list(degrees)]
    top10_nodename = nodename[:10] # geto top 10 degrees, set others to zero
    top10_node_in_deg = node_in_deg[:10] + [0]*(len(node_in_deg)-10) # geto top 10 degrees, set others to zero

    labels = {}    
    for node in G.nodes():
        if node in top10_nodename:
            #set the node name as the key and the label as its value 
            labels[node] = node

 
    pos = nx.random_layout(G)

    plt.figure(figsize=(20,20)) ;

    nx.draw(G,pos,edge_color = 'pink', with_labels = False, nodelist=nodename,
        node_color = 'blue',alpha=0.5,node_size=[20*v for v in node_in_deg],
        width=0.4)
    nx.draw_networkx_labels(G,pos,labels,font_size=20,font_color='black')
    plt.show();
```

## Considering influence based on users based on retweets & favourites & number of mentions & number of followers and following

```python
################################################## dataframe pre-processing
mentions2 = mentions.select(mentions['mentioned_user'],mentions['Num_times_mentioned'])

counts = df.groupBy("username").count().withColumnRenamed("username","username_count")

users = df.groupBy("username","country").agg({'following':'avg', 
                         'followers':'avg', 
                         'retweetcount':'avg', 
                         'favorite_count': 'avg'})
```

```python
############################################### dataframe preprocessing
users_comp = users.join(counts,users.username ==  counts.username_count,"left")\
    .join(mentions2,users.username ==  mentions2.mentioned_user,"left")\
    .drop("username_count")\
    .drop("mentioned_user")\
    .na.fill(value=0)\
    .sort('Num_times_mentioned', ascending=False)\
    .withColumnRenamed("avg(followers)","Followers")\
    .withColumnRenamed("avg(following)","Following")\
    .withColumnRenamed("avg(favorite_count)","Av_favourites")\
    .withColumnRenamed("avg(retweetcount)","Av_retweets")\
    .withColumnRenamed("count","Num_tweets")
    
for c in users_comp.columns[2:]:
    users_comp = users_comp.withColumn(c, round(c).cast('integer'))

users_comp = users_comp.select("username","country","Num_tweets","Followers","Following","Av_retweets","Av_favourites","Num_times_mentioned")
#users_comp.show()
```

```python
#calculating the popularity score
users_comp = users_comp.withColumn('Pop_score',users_comp['Av_retweets'])
#calculating the relevance score
users_comp = users_comp.withColumn('Rel_score',users_comp['Num_times_mentioned'])
#calculating the reach score
users_comp = users_comp.withColumn('Reach_score',users_comp['Followers'] - users_comp['Following'])
#users_comp.show()
```

```python
############################################## taking log transform
users_comp = users_comp.withColumn("Pop_score", log10(col("Pop_score")))
users_comp = users_comp.withColumn("Rel_score", log10(col("Rel_score")))
users_comp = users_comp.withColumn("Reach_score", log10(col("Reach_score")))                                   
```

```python
####################################### filetering to avoid getting users who had just one single tweet with many retweets
users_comp = users_comp.filter("Num_tweets>1")
```

## Calculate overall Normalised score (max score is 3)

```python
############################################## standardising score columns
columns_to_scale = ["Pop_score","Rel_score","Reach_score"]
assemblers = [VectorAssembler(inputCols=[col], outputCol=col + "_vec") for col in columns_to_scale]
scalers = [MinMaxScaler(inputCol=col + "_vec", outputCol=col + "_scaled") for col in columns_to_scale]
pipeline = Pipeline(stages=assemblers + scalers)
scalerModel = pipeline.fit(users_comp)
scaledData = scalerModel.transform(users_comp)
```

```python
# UDF for converting column type from vector to double type
unlist = udf(lambda x: float(list(x)[0]), DoubleType())

scaledData= scaledData.withColumn("Reach_score_scaled", unlist("Reach_score_scaled"))\
.withColumn("Pop_score_scaled", unlist("Pop_score_scaled"))\
.withColumn("Rel_score_scaled", unlist("Rel_score_scaled"))\
.drop(col("Reach_score_vec"))\
.drop(col("Pop_score_vec"))\
.drop(col("Rel_score_vec")) 


#calculating the overall normalised score
scaledData = scaledData.withColumn('Total_score_scaled',scaledData['Reach_score_scaled'] + scaledData['Pop_score_scaled']
                                   + scaledData['Rel_score_scaled'])
```
